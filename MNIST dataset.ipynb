```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
import cv2
```

```python
# Loading MNIST dataset
mnist = tf.keras.datasets.mnist
# This contains both the test and train data, each image of shape - 28X28X1 - grey scale
```

```python
# train dataset - 60,000, test dataset - 10,000
(x_train, y_train),(x_test, y_test) = mnist.load_data()
```

```python
x_train.shape, x_test.shape
```

```python
# let's graph one of the image
plt.imshow(x_train[4], cmap = plt.cm.binary);
# this cmap is to know the change in configuration or colour intensity.
```

### Before Normalization

```python
print(x_train[4])
# near 0 - dark color near 255 - light color
```

##### As images are in gray level (1channel == 0 to 255), not colored (RGB)

### Normalizing the data | Pre-Processing Step

```python
x_train = tf.keras.utils.normalize(x_train, axis = 1)
x_test = tf.keras.utils.normalize(x_test, axis = 1)
plt.imshow(x_train[4], cmap = plt.cm.binary)
```

### After Normalizing

```python
print(x_train[4])
```

```python
print(y_train[4])
```

```python
# Resizing the image to make it suitable for convolution operation

image_size = 28
# increase one dimension for kernal operation
x_train = np.array(x_train).reshape(-1, image_size, image_size, 1)
x_test = np.array(x_test).reshape(-1, image_size, image_size, 1)
print("Training Samples Dimensions", x_train.shape)
print("Testing Samples Dimensions", x_test.shape)
```

### Creating a Deep Neural Network Architecture

```python
# 1)
model = Sequential()

# first convolution layer
model.add(Conv2D(64, (3,3), input_shape = x_train.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

# Second convolution layer
model.add(Conv2D(64, (3,3), input_shape = x_train.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

# Third convolution layer
model.add(Conv2D(64, (3,3), input_shape = x_train.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

# Fourth - FNN-1
model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))

# Fifth - FNN-2
model.add(Dense(32))
model.add(Activation('relu'))

# Sixth - FNN-3
model.add(Dense(10))
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
model.compile(loss = "sparse_categorical_crossentropy", optimizer = "adam", metrics =['accuracy'])
```

```python
# Training the model
model.fit(x_train, y_train, epochs = 5, validation_split = 0.3)
```

```python
# Evaluating on hold-out test dataset
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test Loss on 10,000 samples", test_loss)
print("Test accuracy on 10,000 samples", test_acc)
```

```python
# predict digit using image passed in 
def predict(model, img):
    imgs = np.array([img])
    res = model.predict(imgs)
    index = np.argmax(res)
    #print(index)
    return str(index)

startInference = False
def ifClicked(event, x, y, flags, params):
    global startInference
    if event == cv2.EVENT_LBUTTONDOWN:
        startInference = not startInference

# threshold slider handler 
threshold = 100
def on_threshold(x):
    global threshold
    threshold = x

# the opencv display loop 
def start_cv(model):
    global threshold
    cap = cv2.VideoCapture(1)
    frame = cv2.namedWindow('background')
    cv2.setMouseCallback('background', ifClicked)
    cv2.createTrackbar('threshold', 'background', 150, 255, on_threshold)
    background = np.zeros((480, 640), np.uint8)
    frameCount = 0

    while True:
        ret, frame = cap.read()

        if (startInference):
            
            # frame counter for showing text 
            frameCount += 1

            # black outer frame
            frame[0:480, 0:80] = 0
            frame[0:480, 560:640] = 0
            grayFrame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

            # apply threshold
            _, thr = cv2.threshold(grayFrame, threshold, 255, cv2.THRESH_BINARY_INV)
           

            # get central image 
            resizedFrame = thr[240-75:240+75, 320-75:320+75]
            background[240-75:240+75, 320-75:320+75] = resizedFrame

            # resize for inference 
            iconImg = cv2.resize(resizedFrame, (28, 28))
            
            # pass to model predictor 
            res = predict(model, iconImg)

            # clear background 
            if frameCount == 5:
                background[0:480, 0:80] = 0
                frameCount = 0

            # show text 
            cv2.putText(background, res, (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 3)
            cv2.rectangle(background, (320-80, 240-80), (320+80, 240+80), (255, 255, 255), thickness=3)
            
            # display frame 
            cv2.imshow('background', background)
        else:
            # display normal video 
            cv2.imshow('background', frame)

        # cv2.imshow('resized', resizedFrame)
        if cv2.waitKey(1) & 0xff == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
    
start_cv(model)
```
